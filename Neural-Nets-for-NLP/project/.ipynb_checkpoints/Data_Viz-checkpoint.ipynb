# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

from collections import Counter

from keras.preprocessing.text import Tokenizer

%matplotlib inline  
```

## Read Data

```python
df_text = pd.read_csv('input/PubMed_20k_RCT.csv')
df_postag = pd.read_csv('input/PubMed_20k_RCT_POS_TAG.csv')
df_constree = pd.read_csv('input/PubMed_20k_RCT_CONSTPARSE.csv')
```

## Category breakdown

```python
import matplotlib.pyplot as plt 

plt.rcdefaults()

plt.figure(figsize=(4, 3))

labels = [(label.title(), cnt) for label, cnt in Counter(df_text.label.values).most_common(5)]
 
objects = (labels[0][0], labels[1][0], labels[2][0], labels[3][0], labels[4][0])
y_pos = np.arange(len(objects))
counts = [labels[0][1], labels[1][1], labels[2][1], labels[3][1], labels[4][1]]
 
plt.barh(y_pos, counts, align='center')
plt.yticks(y_pos, objects)
plt.xlabel('Number of labels in dataset')
plt.xticks([0,20000,40000,60000,80000], ['0','20k','40k','60k','80k'])
 
plt.savefig('number_of_labels_in_dataset.png', bbox_inches='tight', pad_inches=0)
```

## Sentence per abstract

```python
# find size of abstracts

# create dataframe with abstract lengths
abstract_lens = df_text[['abstract_id','text']].groupby(by='abstract_id').agg('count')

fig = plt.figure(figsize=(6, 6))

gs = GridSpec(6, 1)
ax1 = plt.subplot(gs[0, :])
ax1.boxplot(abstract_lens.text.values, vert=False, widths=0.4)
ax1.xaxis.tick_top()
ax1.tick_params(left='off',labelleft='off')

ax2 = plt.subplot(gs[1:, :])
ax2.hist(abstract_lens.text.values, bins=40)
ax2.set_xlabel('Number of sentences per abstract', fontsize=12)
ax2.set_ylabel('Number of occurrences', fontsize=12)

#plt.show()
plt.savefig('Sentences_Per_Abstract.png', bbox_inches='tight', pad_inches=0)
```

## Tokens per sentence

```python
X_all = df_text.text.values
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_all)
sequences = tokenizer.texts_to_sequences(X_all)
seq_len = [len(seq) for seq in sequences]
```

```python
fig = plt.figure(figsize=(6, 6))

gs = GridSpec(6, 1)
ax1 = plt.subplot(gs[0, :])
ax1.boxplot(seq_len, vert=False, widths=0.4)
ax1.xaxis.tick_top()
ax1.tick_params(left='off',labelleft='off')

ax2 = plt.subplot(gs[1:, :])
ax2.hist(seq_len, bins=70)
ax2.set_xlabel('Number of tokens per sentence', fontsize=12)
ax2.set_ylabel('Number of occurrences', fontsize=12)

#plt.show()
plt.savefig('Tokens_per_sentence.png', bbox_inches='tight', pad_inches=0)
```

## Constituent parset tree tag per sentence

```python
X_all = df_constree.deptree2.values

tokenizer_ = Tokenizer()
tokenizer.fit_on_texts(X_all)
sequences = tokenizer.texts_to_sequences(X_all)
seq_len = [len(seq) for seq in sequences]
```

```python
fig = plt.figure(figsize=(6, 6))

gs = GridSpec(6, 1)
ax1 = plt.subplot(gs[0, :])
ax1.boxplot(seq_len, vert=False, widths=0.4)
ax1.xaxis.tick_top()
ax1.tick_params(left='off',labelleft='off')

ax2 = plt.subplot(gs[1:, :])
ax2.hist(seq_len, bins=70)
ax2.set_xlabel('Number of constituent parse tree tags per sentence', fontsize=12)
ax2.set_ylabel('Number of occurrences', fontsize=12)

#plt.show()
plt.savefig('constreetags_per_sentence.png', bbox_inches='tight', pad_inches=0)
```
